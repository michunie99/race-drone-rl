```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gateReward(t_pos):
    dp, dn = t_pos[0], t_pos[1]
    # Normalize dn
    #dn /= 0.5
    # TODO - check the cooeficent of 1.5
    f = lambda x: max(1-x/3, 0.0)
    v = lambda x, y: max((1- y) * (x/6.0), 0.05) 
    filed_reward = -f(dp)**2 * (1 - np.exp(-0.5 * dn**2 / v(1.0, f(dp))))
    return filed_reward
```

```python
nx, ny = (100, 100)
x = np.linspace(0, 2, nx)
y = np.linspace(-1, 1, ny)

xv, yv = np.meshgrid(x, y)
zv = np.zeros_like(xv)

for i in range(nx):
    for j in range(ny):
        zv[i,j] = gateReward((xv[i,j], yv[i,j]))
```

```python
plt.contourf(xv, yv, zv, 100)
plt.axis('scaled')
plt.colorbar()
plt.show()
```
